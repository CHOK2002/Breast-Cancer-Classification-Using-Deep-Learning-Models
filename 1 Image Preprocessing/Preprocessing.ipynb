### Image cleaning

Load image and direct filter out the mask image

```python
from PIL import Image
import os
import numpy as np

# Define the folder path
folder_path = os.path.join("Breast Cancer Class", "benign")

# Initialize a list to store non-mask images
benign_images = []

# Loop through all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".png") and "_mask" not in filename:  # Exclude mask files
        file_path = os.path.join(folder_path, filename)
        try:
            # Load the image
            image = Image.open(file_path)
            # Optionally convert to a NumPy array
            image_array = np.array(image)
            # Append the image or its array to the list
            benign_images.append(image_array)
        except Exception as e:
            print(f"Error loading {filename}: {e}")

# Display the number of images loaded
print(f"Loaded {len(benign_images)} non-mask images from {folder_path}")
```

Check duplicated

```python
# Function to check for duplicates in the images list
def check_duplicates(image_list):
    seen = []
    duplicates = []
    for i, img in enumerate(image_list):
        if any(np.array_equal(img, seen_img) for seen_img in seen):
            duplicates.append(i)  # Record the index of the duplicate
        else:
            seen.append(img)  # Add the image to the seen list
    return duplicates

# Check for duplicates in the benign_images list
duplicates = check_duplicates(benign_images)

# Display the result
if duplicates:
    print(f"Found duplicates at indices: {duplicates}")
else:
    print("No duplicates found.")
```

Resize the image

```python
# Initialize a list to store resized images
resized_benign_images = []

# Resize each image in the normal_images list
for img in benign_images:
    # Resize the image to 224x224
    image_resized = Image.fromarray(img).resize((224, 224))
    
    # Convert the image to RGB if it's grayscale (ensuring 3 channels)
    image_resized = image_resized.convert("RGB")
    
    # Optionally convert to a NumPy array
    image_array = np.array(image_resized)
    
    # Append the resized image to the list
    resized_benign_images.append(image_array)

# Display the number of resized images
print(f"Resized {len(resized_benign_images)} images to 224x224x3")
```

Augmentation

```python
import cv2
import numpy as np
import albumentations as A
from matplotlib import pyplot as plt

# Define distinct augmentation transformations
transformations = [
    A.HorizontalFlip(p=1.0),  # Augmentation 1: Horizontal Flip
    # A.RandomRotate90(p=1.0),  # Augmentation 2: 90-degree rotation
    # A.ShiftScaleRotate(scale_limit=0.2, rotate_limit=30, shift_limit=0.1, p=1.0),  # Augmentation 3: Shift/Scale/Rotate
    # A.RandomBrightnessContrast(p=1.0)  # Augmentation 4: Brightness/Contrast adjustment
]

# Initialize the final list to store original and augmented images
final_malignant_images = []

# Assuming resized_cat contains your 5 original resized images
for img in resized_malignant_images:
    # Convert the image to RGB (in case it's in BGR format from OpenCV)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # Add the original image to the final list
    final_malignant_images.append(img)
    
    # Create 4 distinct augmentations for the current image
    for transform in transformations:
        augmented = transform(image=img)
        final_malignant_images.append(augmented['image'])  # Add each augmented image to the final list

# Verify the augmentation process
print(f"Total number of images (original + augmented): {len(final_malignant_images)}")

# Example: Display the first original image and its 4 augmentations
original_image = final_malignant_images[0]  # First original image
augmented_images = final_malignant_images[1:2]  # Its 4 augmentations

# Plot the original image and the 4 augmentations
fig, axes = plt.subplots(1, 2, figsize=(20, 5))  # 1 row, 2 columns (1 original + 1 augmented)

# Display the original image
axes[0].imshow(original_image)
axes[0].set_title("Original Image")
axes[0].axis('off')

# Display the 4 augmented images
for i, augmented_image in enumerate(augmented_images):
    axes[i + 1].imshow(augmented_image)
    axes[i + 1].set_title(f"Augmented {i + 1}")
    axes[i + 1].axis('off')

# Show the plot
plt.tight_layout()
plt.show()
```
