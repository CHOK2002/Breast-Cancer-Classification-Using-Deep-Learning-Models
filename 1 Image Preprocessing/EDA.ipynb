Load and display image

```python
from PIL import Image
import os
import numpy as np

# Define the folder path
folder_path = os.path.join("Breast Cancer Class", "benign")

# Initialize a list to store images
benign_images = []

# Loop through all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".png"):  # Check if the file is a PNG image
        file_path = os.path.join(folder_path, filename)
        try:
            # Load the image
            image = Image.open(file_path)
            # Optionally convert to a NumPy array
            image_array = np.array(image)
            # Append the image or its array to the list
            benign_images.append(image_array)
        except Exception as e:
            print(f"Error loading {filename}: {e}")

# Display the number of images loaded
print(f"Loaded {len(benign_images)} images from {folder_path}")
```

```python
import matplotlib.pyplot as plt

# Check if images are loaded
if len(benign_images) > 0:
    # Display the first 5 images
    for i in range(min(4, len(benign_images))):  # Display up to 5 images if available
        plt.figure(figsize=(4, 4))
        plt.imshow(benign_images[i], cmap='gray')  # Use 'gray' for grayscale images
        plt.title(f"Image {i+1}")
        plt.axis('off')
        plt.show()
else:
    print("No images loaded to display.")
```

Check duplicate image

```python
# Function to check for duplicates in the images list
def check_duplicates(image_list):
    seen = []
    duplicates = []
    for i, img in enumerate(image_list):
        if any(np.array_equal(img, seen_img) for seen_img in seen):
            duplicates.append(i)  # Record the index of the duplicate
        else:
            seen.append(img)  # Add the image to the seen list
    return duplicates

# Check for duplicates in the benign_images list
duplicates = check_duplicates(benign_images)

# Display the result
if duplicates:
    print(f"Found duplicates at indices: {duplicates}")
else:
    print("No duplicates found.")
```

```python
import matplotlib.pyplot as plt

# Function to check for duplicates in the images list
def check_duplicates(image_list):
    seen = []
    duplicates = []
    for i, img in enumerate(image_list):
        if any(np.array_equal(img, seen_img) for seen_img in seen):
            duplicates.append(i)  # Record the index of the duplicate
        else:
            seen.append(img)  # Add the image to the seen list
    return duplicates

# Check for duplicates in the normal_images list
duplicates = check_duplicates(normal_images)

# Display the result and print duplicated images
if duplicates:
    print(f"Found duplicates at indices: {duplicates}")
    # Display the duplicated images
    for idx in duplicates:
        plt.figure(figsize=(5, 5))
        plt.imshow(normal_images[idx])
        plt.title(f"Duplicated Image at index {idx}")
        plt.axis('off')  # Hide axis
        plt.show()
else:
    print("No duplicates found.")
```

Check overall size/shape of image

```python
import pandas as pd

# Group by unique dimensions and calculate counts
image_dimensions = [(img.shape[1], img.shape[0]) for img in benign_images]  # Get width and height
df_dimensions = pd.DataFrame(image_dimensions, columns=['width', 'height'])
df_groupby = df_dimensions.groupby(['width', 'height']).size().reset_index(name='count')

# Plot grouped scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(
    df_groupby['width'], 
    df_groupby['height'], 
    s=df_groupby['count'] * 10,  # Marker size proportional to count
    alpha=0.7, 
    color='blue'
)
plt.title("Scatter Plot of Image Widths and Heights (with Frequency)")
plt.xlabel("Width (pixels)")
plt.ylabel("Height (pixels)")
plt.grid(True)
plt.show()
```
